# **assign data**:
**( a list of exam scores)**

```python
data = [88,85,83,86,85,83,88,88,86,83,85,83,86,85,83,88,88,85]
```

# **Convert the data into a pandas Series**
**pandas series** : a fundamental data structure in the pandas library, designed to represent a single column of data in a tabular form

```python
series_Data = pd.Series(data)
```

# **create a frequency table:**
**value_counts():**function returns a Series with unique values as the index and their respective frequencies as the values.



```python
frequency_table = series_Data.value_counts()
```
